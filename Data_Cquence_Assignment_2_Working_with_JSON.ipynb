```python
import pandas as pd
url = 'https://think.cs.vt.edu/corgis/datasets/json/airlines/airlines.json'
pd.read_json(url).head()
```

```python
import requests
from pandas import json_normalize
datafile = requests.get(url=url)
df = json_normalize(datafile.json())
df.head()
```

```python
 #Find the 5 airports with the most flight delays\n
 df.sort_values('Statistics.Flights.Delayed', ascending=False).head(5)
```

```python
#Find the year with the most number of delays that were caused by Security (using column 'Security')
df.sort_values('Statistics.# of Delays.Security', ascending=False).head(1)
```

```python
#Find the 5 airports with the least flight delays.
df.sort_values('Statistics.Flights.Delayed', ascending=True).head(5)
```
